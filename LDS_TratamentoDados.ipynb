````python
# import libraries
import pandas as pd  # Biblioteca para os DataFrames 
import numpy as np   # Biblioteca Numérica

# Ler o arquivo, e atribua-o à variável "df".
my_file = "https://raw.githubusercontent.com/EduPekUfpr/PythonProject/refs/heads/main/Dados/Auto_limpo.csv"

# Atribuir o arquivo a uma váriável DataFrame
df = pd.read_csv( my_file)

df.head()
````

## Valores Faltantes (Missing Data)

A primeira pergunta a ser feita é quantos dados estão faltando?

### Forma 1:

Para isso vamos criar um DataFrame chamado `missing_data` onde:

* "True" signifca que o valor é faltante 
* "False" signifca que o valor é presente


````python
missing_data = df.isna()
missing_data.head(5)
````

<h3>Contar valores ausentes em cada coluna</h3>
<p>
Usando um loop em Python, podemos descobrir rapidamente o número de valores em falta em cada coluna. Como mencionado acima, "True" representa um valor ausente e "False" significa que o valor está presente no conjunto de dados.  

No corpo do loop `for` , o método ".value_counts()" conta o número de valores "True" e "False". 
</p>

````python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("\n")   
````

### Uma segunda forma mais simplificada de contar os valores faltantes nas colunas

````python
for column in df.columns.values.tolist():
  total = df[column].isna().sum()
  if total > 1:
    print(column, "\t NaN = ",total)
    
````

### Outra forma mais simples ainda

````python
df.isna().sum() # missing_data.sum()
````

### Valores faltantes por linha

* ```df.loc[[i]]```: seleciona a linha de índice i do DataFrame, retornando-a como um DataFrame (não uma Série, por isso o uso de ```[[i]])```.
* ```.isna()```: verifica se há valores faltantes (NA) na linha, retornando uma máscara booleana onde True indica NaN.
* ```.sum()```: a primeira soma conta os valores True (indicando NA) para cada coluna da linha, resultando em uma Série com o total de NaN por coluna.
* A segunda ```.sum()``` soma os valores dessa Série, dando o número total de NaN em toda a linha.

````python
for i in range(df.shape[0]):
  total = df.loc[[i]].isna().sum().sum()
  if total >= 1:
    print("Linha ", i, "\t NaN = ",total)
````

### Outra forma

````python
df.loc[df.isna().sum(axis=1)>=1]
````

#  Quais são as opções para lidar com dados ausentes?


## Deletar os Dados
  * Deletar a linha inteira
  * Deletar a coluna inteira

As colunas inteiras só devem ser descartadas se a maioria dos dados da coluna estiver vazia. Em nosso conjunto de dados, nenhuma das colunas está vazia o suficiente para ser descartada completamente.

Quando temos uma linha com muitos dados faltantes, levanta-se a hipótese da veracidade dela. Portanto também podemos deletá-la.

-----
**Deletar a linha inteira:**

* "preço": 4 dados em falta, basta apagar a linha inteira

Razão: o preço é o que queremos prever. Qualquer entrada de dados sem dados de preço não pode ser usada para previsão; portanto, qualquer linha sem dados de preço não nos é útil.

-----

````python
# Delete a linha toda com NaN na coluna "price" 
df.dropna(subset=["price"], axis=0, inplace = True)

#df = df.dropna(subset=["price"], axis=0)

# reset index, because we droped some rows
df.reset_index(drop = True, inplace = True)

df.head()
````

## Substituir dados
* Substituir pela média
* Substituir pela maior frequência
* Substituir com base em outras funções

Temos alguma liberdade para escolher qual método substituir os dados; no entanto, alguns métodos podem parecer mais razoáveis do que outros. 

**Substituir pela média:**

* "normalized-losses": 41 dados faltantes, substitua-os pela média
*"stroke": 4 dados faltantes, substitua-os pela média
* "bore": 4 dados faltantes, substitua-os pela média<
*"horse-power": 2 dados faltantes, substitua-os pela média
*"peak-rpm": 2 dados faltantes, substitua-os pela média

````python
# Cálculo da média normalized-losses
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)

# Substituição pela média normalized-losses
df["normalized-losses"] = df["normalized-losses"].fillna(avg_norm_loss) 
````

````python
# bore
avg_bore= df["bore"].astype("float").mean(axis=0)
print("Average of bore:", avg_bore)
df["bore"] = df["bore"].fillna(avg_bore) 

# stroke
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average stroke:", avg_stroke)
df['stroke'] = df['stroke'].fillna(avg_stroke)

# horsepower
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'] = df['horsepower'].fillna(avg_horsepower)

# peak-rpm
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'] = df['peak-rpm'].fillna(avg_peakrpm)
````

# Categorical Data

## Valor mais frequente (Dados Categóricos)

**Substituir pela frequência:**

* "num-of-doors": 2 dados faltantes, substitua-os pelo valor mais frequente.

Razão: 84% de sedans são quatro portas. Como quatro portas é a mais frequente, é mais provável que ocorra

````python
# Valor mais frequente de num-of-doors
print(df['num-of-doors'].value_counts()) # imprime a contagem de cada valor

a = df['num-of-doors'].mode() # mode pega valor mais frequente

print("Valor mais frequente de num-of-doors: ", a)

#Substitua o valor 'num-of-doors' pelo mais frequente
df["num-of-doors"] = df["num-of-doors"].fillna(a[0])
````
