```python
import pandas as pd # download library to read data into dataframe
import numpy as np # import numpy library
import matplotlib.pyplot as plt # plots
%matplotlib inline

import collections
import random # library for random number generation

# import decision trees scikit-learn libraries
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix # Não usaremos essas métricas neste exemplo

import matplotlib.pyplot as plt

#!conda install python-graphviz --yes
import graphviz

from sklearn.tree import export_graphviz

import itertools
```

```python
file_path = "https://raw.githubusercontent.com/EduPekUfpr/PythonProject/refs/heads/main/Dados/H_M_carac.csv"

HM_data = pd.read_csv(file_path)
HM_data.head()
```

```python
X = HM_data.drop(['sexo'], axis = 1)
y = HM_data['sexo']

classification_tree = tree.DecisionTreeClassifier(max_depth = 4, min_samples_split = 2, min_samples_leaf = 1)
classification_tree.fit(X, y)
```
