```python
# ==========================================================
# TRATAMENTO DE VALORES AUSENTES NUMÉRICOS
# Impacto nas Estatísticas Descritivas
# ==========================================================

import pandas as pd
import numpy as np

# ==========================================================
# 1. CRIANDO UM DATASET EXEMPLO
# ==========================================================

np.random.seed(42)
dados = {
    "idade": [23, 25, 29, np.nan, 31, 35, np.nan, 40, 42, 50],
    "salario": [2500, 3200, np.nan, 4100, 5000, np.nan, 6200, 7000, 8000, 15000]
}

df_original = pd.DataFrame(dados)
print("DATASET ORIGINAL")
print(df_original)

# ==========================================================
# 2. FUNÇÃO PARA MOSTRAR ESTATÍSTICAS
# ==========================================================

def mostrar_estatisticas(df, titulo):

    print("\n" + "="*60)
    print(titulo)
    print("="*60)
    print("\nMÉDIA")
    print(df.mean(numeric_only=True))
    print("\nMEDIANA")
    print(df.median(numeric_only=True))
    print("\nDESVIO PADRÃO")
    print(df.std(numeric_only=True))
    print("\nMÍNIMO")
    print(df.min(numeric_only=True))
    print("\nMÁXIMO")
    print(df.max(numeric_only=True))
    print("\nQUANTIDADE DE LINHAS")
    print(len(df))
```

```python
# ==========================================================
# 3. ESTATÍSTICAS ORIGINAIS
# ==========================================================

mostrar_estatisticas(
    df_original,
    "ESTATÍSTICAS ORIGINAIS"
)
```

```python
# ==========================================================
# 4. SUBSTITUIÇÃO POR ZERO
# ==========================================================

df_zero = df_original.fillna(0)
print("\n\nDATASET - SUBSTITUIÇÃO POR ZERO")
print(df_zero)
mostrar_estatisticas(
    df_zero,
    "SUBSTITUIÇÃO POR ZERO"
)
```

```python
# ==========================================================
# 5. SUBSTITUIÇÃO PELA MÉDIA
# ==========================================================

df_media = df_original.copy()
for coluna in df_media.columns:
    media = df_media[coluna].mean()
    df_media[coluna] = df_media[coluna].fillna(media)

print("\n\nDATASET - SUBSTITUIÇÃO PELA MÉDIA")
print(df_media)
mostrar_estatisticas(
    df_media,
    "SUBSTITUIÇÃO PELA MÉDIA"
)
```

```python
# ==========================================================
# 6. FORWARD FILL (VALOR ANTERIOR)
# ==========================================================

df_ffill = df_original.fillna(method="ffill")
print("\n\nDATASET - FORWARD FILL")
print(df_ffill)
mostrar_estatisticas(
    df_ffill,
    "FORWARD FILL"
)
```

```python
# ==========================================================
# 7. ELIMINAÇÃO DE LINHAS
# ==========================================================

df_drop_linhas = df_original.dropna()
print("\n\nDATASET - ELIMINAÇÃO DE LINHAS")
print(df_drop_linhas)

mostrar_estatisticas(
    df_drop_linhas,
    "ELIMINAÇÃO DE LINHAS"
)
```

```python
# ==========================================================
# 8. ELIMINAÇÃO DE COLUNAS
# ==========================================================

df_drop_coluna = df_original.drop(columns=["salario"])
print("\n\nDATASET - ELIMINAÇÃO DE COLUNA")
print(df_drop_coluna)

mostrar_estatisticas(
    df_drop_coluna,
    "ELIMINAÇÃO DA COLUNA 'salario'"
)
```

```python
# ==========================================================
# 9. COMPARAÇÃO DAS Estatísticas
# ==========================================================

comparacaoAverage = pd.DataFrame({
    "Original": df_original.mean(numeric_only=True),
    "Zero": df_zero.mean(numeric_only=True),
    "Média": df_media.mean(numeric_only=True),
    "Forward Fill": df_ffill.mean(numeric_only=True),
    "Drop Linhas": df_drop_linhas.mean(numeric_only=True)
})

print("\n\nCOMPARAÇÃO DAS MÉDIAS")
print(comparacaoAverage.round(2))

comparacaoStDev = pd.DataFrame({
    "Original": df_original.std(numeric_only=True),
    "Zero": df_zero.std(numeric_only=True),
    "Média": df_media.std(numeric_only=True),
    "Forward Fill": df_ffill.std(numeric_only=True),
    "Drop Linhas": df_drop_linhas.std(numeric_only=True)
})

print("\n\nCOMPARAÇÃO Dos Desvios-Padrão")
print(comparacaoStDev.round(2))
```
