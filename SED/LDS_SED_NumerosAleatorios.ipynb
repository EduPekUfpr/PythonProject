# Gerador de números aleatórios

Tempo estimado: **10** minutos

## Objetivos

Após completar esta aula, você será capaz de:

* Entender a lógica do Gerador Congruencial Linear (LCG)
* Usar o numpy para gerar números aleatórios
---


```python
def lcg(seed, a, c, m, n):
    """
    seed: valor inicial (X0)
    a: multiplicador
    c: incremento
    m: módulo
    n: quantidade de números a gerar
    """

    X = seed

    numeros = []
    for _ in range(n):

        X = (a * X + c) % m
        numeros.append(X / m)  # normaliza para (0,1)

    return numeros

# Parâmetros exemplo

seed = 42

a = 1664525

c = 1013904223

m = 2**32

# Gerar 10 números

numeros = lcg(seed, a, c, m, 10)

print(numeros)
```

## Repeticão aparece quando usamos valores pequenos

```python
numeros = lcg(seed, 17, 11, 7, 10)

print(numeros)
```

## Histograma

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeros = lcg(seed=42, a=1664525, c=1013904223, m=2**32, n=1000)

sns.histplot(x=numeros, bins=30)

plt.title("Distribuição dos números gerados (LCG)")
plt.xlabel("Valor")
plt.ylabel("Frequência")

plt.show()
```
