* Verificar a convergência de uma distribuição qualquer para a Normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# ==========================================
# Teorema Central do Limite + Normalidade
# ==========================================

# Semente
np.random.seed(42)

# População NÃO normal (exponencial)
populacao = np.random.exponential(scale=10, size=100000)
#populacao = np.random.gamma(shape = 1, scale=10, size=100000)
#populacao = np.random.uniform(low=0.0, high=50.0, size=100000)
#populacao = np.random.poisson(lam=1.0, size=100000)


# Tamanhos das amostras
tamanhos_amostra = [2, 5, 30, 50, 100]

# Número de amostras
num_amostras = 5000

# ==========================================
# População original
# ==========================================

plt.figure(figsize=(8,4))
plt.hist(populacao, bins=60, density=True, edgecolor='black')
plt.title("Distribuição da População Original")
plt.xlabel("Valor")
plt.ylabel("Densidade")
plt.show()

# ==========================================
# Médias amostrais
# ==========================================

for n in tamanhos_amostra:

    medias = []

    # gera várias amostras
    for i in range(num_amostras):

        amostra = np.random.choice(populacao,
                                   size=n,
                                   replace=True)

        medias.append(np.mean(amostra))

    medias = np.array(medias)

    # ==========================
    # Teste de normalidade
    # ==========================

    # Shapiro-Wilk
    stat, p = shapiro(medias[:500])

    print("="*50)
    print(f"Tamanho da amostra: n = {n}")
    print(f"Estatística Shapiro-Wilk: {stat:.4f}")
    print(f"p-value: {p:.6f}")

    if p > 0.05:
        print("Conclusão: distribuição aproximadamente NORMAL")
    else:
        print("Conclusão: distribuição NÃO normal")

    # ==========================
    # Histograma
    # ==========================

    plt.figure(figsize=(8,4))

    plt.hist(medias,
             bins=50,
             density=True,
             edgecolor='black')

    plt.title(f"Médias Amostrais — n = {n}")
    plt.xlabel("Média")
    plt.ylabel("Densidade")

    plt.show()
```
